# **Generating Images by blending**

```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a folder on the desktop to save the synthetic images
output_folder = os.path.expanduser("~/Desktop/Synthetic_MiddleCenter")
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Function to blend MiddleLeft and MiddleRight images
def blend_images(left_image, right_image):
    # Resize images to the same dimensions (640x480)
    left_image = cv2.resize(left_image, (640, 480))
    right_image = cv2.resize(right_image, (640, 480))

    # Blend the images by averaging pixel values (for simplicity)
    blended_image = cv2.addWeighted(left_image, 0.5, right_image, 0.5, 0)

    return blended_image

# Function to generate synthetic "MiddleCenter" images
def generate_synthetic_middle_center(my_dict, num_images, save_path):
    synthetic_images = []

    # Get the 'MiddleLeft' and 'MiddleRight' image lists from the dictionary
    middle_left_images = my_dict.get("MiddleLeft", [])
    middle_right_images = my_dict.get("MiddleRight", [])

    # Check if the lists for MiddleLeft and MiddleRight are empty
    if not middle_left_images or not middle_right_images:
        print("Error: 'MiddleLeft' or 'MiddleRight' labels are missing in the dataset.")
        return synthetic_images  # Exit if no images are found

    for i in range(num_images):
        # Select random images from 'MiddleLeft' and 'MiddleRight'
        left_img_name = random.choice(middle_left_images)
        right_img_name = random.choice(middle_right_images)

        # Construct the full image paths
        left_img_path = os.path.join('/content/ImprovementSet/ImprovementSet', "MiddleLeft", left_img_name)
        right_img_path = os.path.join('/content/ImprovementSet/ImprovementSet', "MiddleRight", right_img_name)

        # Read the images using cv2.imread
        left_img = cv2.imread(left_img_path)
        right_img = cv2.imread(right_img_path)

        # Blend the images to create a synthetic "Middle Center"
        synthetic_image = blend_images(left_img, right_img)

        # Save the generated image
        output_path = os.path.join(save_path, f"synthetic_middle_center_{i+1}.jpg")
        cv2.imwrite(output_path, synthetic_image)

        synthetic_images.append(synthetic_image)

    return synthetic_images

# Function to plot images
def plot_images(images, num_images_to_plot):
    fig, axes = plt.subplots(1, num_images_to_plot, figsize=(15, 5))
    for i in range(num_images_to_plot):
        ax = axes[i]
        ax.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
        ax.axis('off')  # Turn off axis
        ax.set_title(f"Image {i+1}")
    plt.show()

# Call the function to generate synthetic data
num_images_to_generate = int(input("Enter the number of synthetic images to generate: "))
synthetic_images = generate_synthetic_middle_center(my_dict, num_images_to_generate, output_folder)

# Plot the generated synthetic images
if synthetic_images:
    plot_images(synthetic_images, num_images_to_generate)
    print(f"Generated and displayed {len(synthetic_images)} synthetic images.")
else:
    print("No synthetic images were generated.")
```

# **Generating using pupil shifting**

```python
import cv2
import numpy as np
import random
import os
import matplotlib.pyplot as plt

# Create a folder on the desktop to save the synthetic images
output_folder = os.path.expanduser("~/Desktop/Synthetic_MiddleCenter")
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Function to detect pupil using HoughCircles
def detect_pupil(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)

    # Use HoughCircles to detect circles (pupils)
    circles = cv2.HoughCircles(
        gray,
        cv2.HOUGH_GRADIENT, dp=1.2, minDist=30,
        param1=50, param2=30, minRadius=15, maxRadius=30
    )

    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        return circles[0]  # Return the first circle detected (pupil)
    return None  # If no pupil is detected

# Function to shift the pupil towards the center of the image
def shift_pupil(image, pupil_coordinates):
    h, w = image.shape[:2]

    # Calculate the center of the image
    center_x, center_y = w // 2, h // 2

    # Calculate the shift needed for the pupil
    shift_x = center_x - pupil_coordinates[0]
    shift_y = center_y - pupil_coordinates[1]

    # Translate the image to move the pupil towards the center
    M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
    shifted_image = cv2.warpAffine(image, M, (w, h))

    return shifted_image

# Function to generate synthetic "MiddleCenter" images
def generate_synthetic_middle_center(my_dict, num_images, save_path):
    synthetic_images = []

    # Get the 'MiddleLeft' and 'MiddleRight' image lists from the dictionary
    middle_left_images = my_dict.get("MiddleLeft", [])
    middle_right_images = my_dict.get("MiddleRight", [])

    # Check if the lists for MiddleLeft and MiddleRight are empty
    if not middle_left_images or not middle_right_images:
        print("Error: 'MiddleLeft' or 'MiddleRight' labels are missing in the dataset.")
        return synthetic_images  # Exit if no images are found

    for i in range(num_images):
        # Select random images from 'MiddleLeft' and 'MiddleRight'
        left_img_name = random.choice(middle_left_images)
        right_img_name = random.choice(middle_right_images)

        # Construct the full image paths
        left_img_path = os.path.join('/content/ImprovementSet/ImprovementSet', "MiddleLeft", left_img_name)
        right_img_path = os.path.join('/content/ImprovementSet/ImprovementSet', "MiddleRight", right_img_name)

        # Read the images using cv2.imread
        left_img = cv2.imread(left_img_path)
        right_img = cv2.imread(right_img_path)

        # Detect the pupil in both images
        left_pupil = detect_pupil(left_img)
        right_pupil = detect_pupil(right_img)

        if left_pupil is not None and right_pupil is not None:
            # Shift both images' pupils towards the center
            left_img_centered = shift_pupil(left_img, left_pupil[:2])
            right_img_centered = shift_pupil(right_img, right_pupil[:2])

            # Stack the images side by side to create the "MiddleCenter"
            synthetic_image = np.hstack((left_img_centered, right_img_centered))

            # Save the generated image
            output_path = os.path.join(save_path, f"synthetic_middle_center_{i+1}.jpg")
            cv2.imwrite(output_path, synthetic_image)

            synthetic_images.append(synthetic_image)

    return synthetic_images

# Function to plot images
def plot_images(images, num_images_to_plot):
    fig, axes = plt.subplots(1, num_images_to_plot, figsize=(15, 5))
    for i in range(num_images_to_plot):
        ax = axes[i]
        ax.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for displaying
        ax.axis('off')  # Turn off axis
        ax.set_title(f"Image {i+1}")
    plt.show()

# Call the function to generate synthetic data
num_images_to_generate = int(input("Enter the number of synthetic images to generate: "))
synthetic_images = generate_synthetic_middle_center(my_dict, num_images_to_generate, output_folder)

# Plot the generated synthetic images
if synthetic_images:
    plot_images(synthetic_images, num_images_to_generate)
    print(f"Generated and displayed {len(synthetic_images)} synthetic images.")
else:
    print("No synthetic images were generated.")
```

# ***Resnet***

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# ================================
# Parameters and Dataset Setup
# ================================

# Path to your dataset folder (organized by class)
data_dir = '/content/ImprovementSet/ImprovementSet'

# Image size and batch size
img_size = (224, 224)
batch_size = 32

# Create training and validation datasets using a 70/30 split.
# (You can further split the validation set into cross-validation/test if needed.)
train_ds = image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset='training',
    seed=42,
    image_size=img_size,
    batch_size=batch_size
)
val_ds = image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset='validation',
    seed=42,
    image_size=img_size,
    batch_size=batch_size
)


print("Class names:", train_ds.class_names)

# Apply ResNet50 preprocessing to the datasets
def preprocess(image, label):
    image = preprocess_input(image)
    return image, label

train_ds = train_ds.map(preprocess)
val_ds = val_ds.map(preprocess)

# ================================
# Model Building with ResNet50
# ================================

# Load the ResNet50 base model pre-trained on ImageNet (exclude top layers)
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_size[0], img_size[1], 3))
base_model.trainable = False  # Freeze the base model to focus on training the new top layers

# Create a new model on top of the ResNet50 base
inputs = tf.keras.Input(shape=(img_size[0], img_size[1], 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation='relu')(x)
# Final layer for 8 classes (distracted categories)
outputs = layers.Dense(8, activation='softmax')(x)
model = tf.keras.Model(inputs, outputs)

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])


resnet_checkpoint = ModelCheckpoint('/content/resent_best_model.h5', save_best_only=True, mode='max')
# ================================
# Model Training
# ================================


epochs = 5  # Increase this for better performance
history = model.fit(train_ds, validation_data=val_ds, epochs=epochs, callbacks=[resnet_checkpoint])

# Evaluate on the validation dataset
val_loss, val_acc = model.evaluate(val_ds)
print(f'Validation accuracy: {val_acc:.4f}')

# ================================
# Inference with Threshold for "Focused" Detection
# ================================
```

```python
def classify_image(image_path, threshold=0.6):
    """
    Load an image from the given path, preprocess it, and predict the class probabilities.
    If the maximum probability is below the threshold, return "focused"; otherwise return the predicted class.
    """
    # Load and preprocess the image
    img = tf.keras.preprocessing.image.load_img(image_path, target_size=img_size)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)

    # Predict probabilities for the 8 classes
    preds = model.predict(img_array)[0]
    max_prob = np.max(preds)
    predicted_class = np.argmax(preds)

    # If the highest probability is below the threshold, consider the image as "focused"
    if max_prob < threshold:
        return "focused", preds
    else:
        # Retrieve the class name from the training dataset class names
        class_name = train_ds.class_names[predicted_class]
        return class_name, preds


example_image_path = '/content/ImprovementSet/ImprovementSet/TopRight/9723.jpg'
result, probabilities = classify_image(example_image_path, threshold=0.6)
print(f'Predicted result: {result}')
print(f'Predicted probabilities: {probabilities}')

# Optionally, display the image and predicted probabilities
img = tf.keras.preprocessing.image.load_img(example_image_path, target_size=img_size)
plt.imshow(img)
plt.title(f'Prediction: {result}')
plt.axis('off')
plt.show()
```
